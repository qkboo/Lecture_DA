### 최빈값 mode

최빈값(mode)은 데이터에서 가장 빈번하게 출현하는 값이다. 그래서 공산품 같은 경우 최빈값을 이용해야 더 많은 사람들에 맞는 제품을 만들 수 있다.

```
np.bincount(X)
 - X: 1차워 배열
```` 

np.bincount()로 발생하는 숫자의 빗도수를 얻어 최빈값을 파악할 수 있다. 

예를 들어 한 반에 학생 20명의 수학시험 점수 - 만점 10이 있다. 여기서 빈번히 발생하는 점수는?

`````python
#
np.bincount(scores)
`````


# 데이터가 흩어진 정도 조사하기

일일 교통량을 측정했더니 평균이 20이라고 한다. 평균 20이라는 것은 어떤 날은 교통량이 18대, 또 어떤 날은 19대, ..., 20대, 24대, 21대, 20대, ... 이런 식으로 날마다 조금씩 다를 수는 있지만 평균적으로 20대라는 의미이다.

편차는 평균과의 차이다. (편차 = 데이터 값 - 평균값) 즉 일일 교통량의 표준편차가 3이라는 것은 편차가 평균적으로 3대 정도 된다는 의미이다. 위 예에서 보면, 편차는 -2대, -1대, ..., 0대, 4대, 1대, 0대, ... 이런 식이다. 

편차들을 곧이 곧대로 평균을 해 버리면 +와 -가 상쇄되어 버리기 때문에 항상 0이 나온다. 그래서 제곱을 해서 평균을 취한 것이 분산이다. 즉, `-2*-2, -1*-1, ..., 0*0, 4*4, 1*1, 0*0,...` 이렇게 편차들을 제곱한 후에 얘네들의 평균을 취한 것이 분산이다.

그리고, 이렇게 구한 분산에다가 다시 루트(root)를 취한 것이 표준편차이다

### 분산과 표준편차


분산과 편차는 데이터가 흩어진 정도이다. 이 두 가지 값이 클수록 집단의 데이터가 흩어져 있음을 나타낸다. 특히 편차는 평균에서 *멀어진 정도*라고 이해하면 된다.

**분산 Veriance**는 개별 요소들이 평균과 얼마나 떨어저 있는지 알 수 있다. **편차 Deviation**은 개별 값과 평균의 차이이다.

### 표준편차

산포도에서 편차를 표준편차로 사용한다. 표준편차를 쓰는 이유는, **데이터의 단위에 관계없이 중심경향성은 같은 단위를 가져야** 한다. 산포도 범위range 또한 같은 단위이다. 하지만 분산의 단위는 제곱이다. 그렇기 때문에 분산 대신 원래 단위와 같은 단위를 가지는 ***표준편차 Standard Deviation** 을 이용할 때가 많다.

> 예를 들어 달러로 모은 기부금들의 분산은 제곱 달러로 표시해야 하는데 표준편차는 달러를 그대로 사용할 수 있는 것이다.


일일 교통량의 평균이 20이고 표준편차가 3이라는 것은, 평균적으로 하루에 20대 정도 차가 지나가는데 들쑥 날쑥한 정도가 평균적으로 3대 정도라는 의미이다. 즉, 17대가 지나갈 때도 있고 23대가 지나갈 때도 있다는 것이다.


또한, 일일 교통량의 평균이 80이고 표준편차가 3이라는 것은, 평균적으로 하루에 80대 정도 차가 지나가는데 들쑥 날쑥한 정도가 평균적으로 3대 정도라는 의미이다. 즉, 77대가 지나갈 때도 있고 83대가 지나갈 때도 있다는 것이다.


표준편차 Standard Variance 는 샘플 분산의 양의 제곱근 값이다. s라고 표기한다.

$$ s = \sqrt{s^2}$$

### 마할라노비스 거리(Mahalanobis distance)

마할라노비스 거리는 평균과의 거리가 표준편차의 몇 배인지를 나타내는 값이다.  어떤 값이 얼마나 일어나기 힘든 값인지, 또는 얼마나 이상한 값인지를 수치화하는 한 방법이다. 표준편차로 평균과 차이를 나눈 값이다.

위 예에서(하루 교통량 평균이 20대이고 표준편차가 3대) 어느 날 차가 26대가 지나갔다고 하자. 자로 잰 평균과의 거리는 6이다. 그런데, 마할라노비스 거리로는 (26 - 20)/3 = 2이다. 즉, 표준적인 편차의 2배 정도의 오차가 있는 값이라는 것이다.

1년 내내 매일 매일 차가 정확히 20대만 지나갔었는데 어느날 보니 차가 21대가 지나갔다. 얼마나 이상한 일인가? 이 경우 마할노비스 거리는 굉장히 큰 값을 가질 것이다.  그런데, 차가 어느날은 10대, 다음날은 30대, 또 다른날은 24대, ... 이와 같이 들쑥 날쑥한 경우에 21대가 지나간 것은 전혀 이상한 일이 아닐 것이다. 그래서 이 경우 Mahalanobis distance는 굉장히 작은 값이 나올 것이다.

Mahalanobis distance는 어떤 데이터가 가짜 데이터인지, 아니면 진짜 데이터인지를 구분하는 용도로 주로 사용된다. 

우리가 고등학교 수학에서 (확률과 통계 파트) 표준정규분포로 바꾼 후에 z 값을 구하는 것이 바로 Mahalanobis distance를 구하는 과정이다. 그리고 우리가 통상적으로 사용하는 자로 잰 거리는 유클리디언 거리(Euclidean distance)라고 부른다. 

마할라노비스 거리는 데이터 밀도(분산)를 고려한 거리로 아래 그림을 보면 검은색 점와 타원 경계의 점들 사이의 마할라노비스 거리는 같게 됩니다.

 https://blog.naver.com/pmw9440/222158927060

(X-Y)\sum^{-1}(X-Y)
\right]^{\frac{1}{2}}

## 산포도



데이터 속성 사이의 관계를 파악하는데 **산포도(산점도)**로 양.음의 관계를 파악한다.

2변량 (혹은 이상)의 데이터의 경향을 파악하는데 산점도(산포도) 그래프가 제격이다. 다음 같은 어느 학생들의 수학, 물리 점수이다.




수학 점수를 가로축, 물리 점수를 세로축인 좌표로 생각해 보면, 1번 학생은 (50, 40) 좌표라 할 수 있다.

`````python
score_math = [50, 60, 40, 30, 70, 50]
score_pysic = [40, 60, 40, 20, 80, 50]


plt.figure(figsize=(4,4))
plt.xlabel('수학')
plt.ylabel('물리')
plt.scatter(score_math, score_pysic,s=100, marker='.')
plt.axis('equal')

plt.show()
`````

산포도 각 점은 우상향의 띄를 보여주는데 이것은 양의 기울기를 갖는 1차함수 그래프를 닮았다. 이럴때 통계에서는 두 속성 사이에 강한 양의 상관관계가 있다고 한다.

2변량 데이터의 산포도로 두 변량(속성) 사이의 상관관계 유무의 강.약을 알 수 있다.

기온이 27도를 넘으면 아이스크림이 잘 팔린다. 사마귀가 높은 곳에 알을 낳으면 그 해는 적설량이 많다 등은 모두 과거 통계 데이터를 토대로한 이야기 이다. 이것을 이용해 미래를 예측해보고자 한다.

어떤 주스 판매량을 기반으로한 산포도 그래프이다. 기온이 높아지면 주스 판매량이 늘어나는 것이 보이지만 얼마나 늘어나는지 확인을 위해 점 집단 사이에 경향을 잘 나타내는 직선을 그려보면 눈 대중으로 그리다 보니 아래 그림 처럼 기울기가 들쑥 날쑥 할 수 있다.


[파이썬: 다시 배우는 핵심 고등 수학]

이상적인 것은 각 점에 대해서 오차가 작게 그려지는 것이다. 이것은 선과 모든 점의 오차를 더해 결과가 최소가 되도록 해야 하는데 이를 위해 직선의 기울기와 절편을 구하면 된다. 그런데 점과 오차에는 양수, 음수가 섞여 있으므로 오차를 모두 제곱한 후에 더한다 - 이를 회귀분석이라 한다.

이를 통해 가장 오차가 작게 나오는 직선을 **회귀 직선**이라고 한다.


[파이썬: 다시 배우는 핵심 고등 수학]

#### numpy 회귀직선 기울기, 절편

[numpy.polyfit()](https://docs.scipy.org/doc/numpy/reference/generated/numpy.polyfit.html) 함수

```python
numpy.polyfit(x, y, deg, rcond=None, full=False, w=None, cov=False)
 - deg: 차원.
```

회귀직선 그리기

`````python
# 데이터
x = np.array([23,24,28,24,27,21,18,25,28,20])  # 기온
y = np.array([37,22,62,32,74,16,10,69,83,7])   # 쥬스의 판매량

# 회귀직선
a, b = np.polyfit(x, y, 1)
y2 = a * x + b
print('기울기: {0}, 절편:{1}'.format(a, b))
`````

`````python
# 그린다

plt.scatter(x, y)  # 산포도
plt.plot(x, y2)    # 회귀직선
plt.show()
`````

### where()

> where(조건, 참데이터, 거짓데이터)

조건을 만족하면 참, 거짓 데이터 사용

`````python
xData = np.array([1.1,1.2,1.3,1.4,1.5])
yData = np.array([2.1,2.2,2.3,2.4,2.5])
`````

`````python
result = np.where([True,False,True,True,False], xData, yData)
result
`````

`````python
# 0보다 크면 2, 작으면 -2
testData = np.random.randn(4,4) # 4x4
result = np.where( testData > 0, 2, -2) # 참이면 2, 거짓이면 -2
result
`````

`````python
# 0보다 크면 2, 작으면 그대로
np.where( testData > 0, 2, testData)
`````

`````python

# 특정 조건을 만족하는 총 개수
(testData > 0).sum()
`````

`````python
#
testData.all()
`````

`````python


testData.any()
`````
