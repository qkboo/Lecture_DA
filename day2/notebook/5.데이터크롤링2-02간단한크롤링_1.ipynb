# 네이버 영화 랭킹

```python
import requests
from bs4 import BeautifulSoup
```

https://movie.naver.com/

```python
response = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.naver')
response.status_code
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
soup.find_all('table')
```

```python
soup.select('#old_content > table > tbody > tr:nth-child(2) > td.title > div')
```

```python
soup.select('div[class=tit3]')
```

```python
tags = soup.select('div[class=tit3]')
for tag in tags:
    print(tag.text.strip())  # strip() 으로 공백 제거
```
